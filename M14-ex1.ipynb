```python
import pandas as pd
import matplotlib.pyplot as plt

# Lista de nomes de arquivos CSV
nomes_arquivos = [
    'SINASC_RO_2019_MAR.csv',
    'SINASC_RO_2019_ABR.csv',
    'SINASC_RO_2019_MAI.csv',
    'SINASC_RO_2019_JUN.csv',
    'SINASC_RO_2019_DEZ.csv'
]

# Inicializa um DataFrame vazio para armazenar os dados combinados
dados_combinados = pd.DataFrame()

# Loop através dos nomes de arquivos e carrega os dados em um único DataFrame
for nome_arquivo in nomes_arquivos:
    caminho_arquivo = nome_arquivo  # Pode ser necessário especificar o caminho completo do arquivo
    dados_arquivo = pd.read_csv(caminho_arquivo)
    dados_combinados = pd.concat([dados_combinados, dados_arquivo], ignore_index=True)

# Exibe as primeiras linhas do DataFrame combinado
dados_combinados.head()

# Extrair o mês a partir da coluna 'DTNASC'
dados_combinados['Mês'] = pd.to_datetime(dados_combinados['DTNASC'], format='%Y-%m-%d').dt.strftime('%b')

# Filtrar os dados para os meses de março, abril, maio, junho e dezembro
meses_selecionados = ['Mar', 'Apr', 'May', 'Jun', 'Dec']
dados_filtrados = dados_combinados[dados_combinados['Mês'].isin(meses_selecionados)]

# Calcular a contagem de ocorrências para cada mês
contagem_por_mes = dados_filtrados['Mês'].value_counts().sort_index()

# Criar um gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(contagem_por_mes.index, contagem_por_mes.values)
plt.xlabel('Mês')
plt.ylabel('Contagem')
plt.title('Contagem de Ocorrências por Mês')
plt.show()

```
